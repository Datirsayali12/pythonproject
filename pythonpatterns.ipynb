```python
a = 153
original_num = a
sum = 0
n = len(str(a))

while a != 0:
    p = a % 10
    sum = sum + p**n
    a = a // 10

if sum == original_num:
    print('Armstrong')
else:
    print('Not Armstrong')
```

# hollow pattern:

```python
n=5
for i in range(n):
    for j in range(n):
        if i==0 or i==n-1 or j==0 or j==n-1:
            print('*',end=' ')
        else:
            print(' ',end=' ')
    print()
```
